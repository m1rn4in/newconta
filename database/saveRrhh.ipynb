```python
import pandas as pd
from database.client import get_database
from pymongo.database import Database
```

```python
db:Database = get_database()
```

```python
topics_n1_rrhh = pd.read_excel("./data-bot-rrhh.xlsx",sheet_name="TopicosN1")
topics_n2_rrhh = pd.read_excel("./data-bot-rrhh.xlsx",sheet_name="TopicosN2")
topics_n3_rrhh = pd.read_excel("./data-bot-rrhh.xlsx",sheet_name="TopicosN3")
```

```python
topics_n1_rrhh.head()
```

```python
data_list = []
for i in range(len(topics_n1_rrhh)):
    data_list.append(topics_n1_rrhh.iloc[i].to_dict())
data_list
```

```python
collection = db["TopicosN1Rrhh"]
collection.insert_many(data_list)
```

```python
topics_n1_in_db_rrhh = list(db["TopicosN1Rrhh"].find())
for topic in topics_n1_in_db_rrhh:
    print(topic)
```

```python
topics_n2_rrhh.head()
```

```python
data_list = []
for topicn1 in topics_n1_in_db_rrhh:
    topic_n1_data  = topics_n2_rrhh[topics_n2_rrhh["T1"]==topicn1["T1"]]
    
    for  i in range(len(topic_n1_data)):
        topic_n2_data = topic_n1_data.iloc[i].to_dict()
        topic_n2_data["T1"] = topicn1["_id"]
        data_list.append(topic_n2_data)
```

```python
collection = db["TopicosN2Rrhh"]
collection.insert_many(data_list)
```

```python
topics_n2_in_db_rrhh = list(db["TopicosN2Rrhh"].find())
for topic in topics_n2_in_db_rrhh:
    print(topic)
```

```python
topics_n3_rrhh.head()
```

```python
data_list = []
for topicn2 in topics_n2_in_db_rrhh:
    topic_n2_data  = topics_n3_rrhh[topics_n3_rrhh["T2"]==topicn2["T2"]]
    
    for  i in range(len(topic_n2_data)):
        topic_n3_data = topic_n2_data.iloc[i].to_dict()
        del topic_n3_data["T1"]
        topic_n3_data["T2"] = topicn2["_id"]
        data_list.append(topic_n3_data)
```

```python
collection = db["TopicosN3Rrhh"]
collection.insert_many(data_list)
```

```python
topics_n3_in_db = list(db["TopicosN3Rrhh"].find())
for topic in topics_n3_in_db:
    print(topic)
```
